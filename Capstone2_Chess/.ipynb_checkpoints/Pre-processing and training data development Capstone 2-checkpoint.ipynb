```python
#Importing some packages I'm confident I'll be using
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.preprocessing
```

### Goal: Create a cleaned development dataset you can use to complete the
### modeling step of your project. 


```python
#let's load in the most popular dataset
df = pd.read_csv('most_popular_over_200_games.csv', index_col = False)
df.drop(columns = 'game_id', inplace = True)
df.head()
```

```python
#a little worreid about how long this file is going to get, but let's go ahead and use get_dummies on the opening_name column
dummy = pd.get_dummies(df['opening_name'])
#I've checked this out, but then got errors later on because of the columns still using string objects, so let's dummy the entire df
full_dummy = pd.get_dummies(df)
full_dummy.head()
```

```python
#doing some preprocessing, importing a few different scalers to try different routes
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(full_dummy)
new_df = pd.DataFrame(df_scaled, columns = full_dummy.columns)
new_df.head()
```

```python
#train test split time
from sklearn.model_selection import train_test_split
train, test = train_test_split(new_df, test_size = 0.3)
print(train.shape, test.shape)
```

```python
#and let's go ahead and save it
train.to_csv('training_set_standard_scaler.csv')
test.to_csv('test_set_standard_scaler.csv')
```
