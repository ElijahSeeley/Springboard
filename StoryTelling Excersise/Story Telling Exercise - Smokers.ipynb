```python
#adding expected packages
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#opening Smoker's dataset
df = pd.read_csv('tobacco.csv')
df.head()
df.drop(columns = 'Location 1', inplace = True)
```

```python
new_york = df.loc[df['State'] == 'New York']
new_york.sort_values(by ='Year', inplace = True)
new_york.head()
```

```python
new_york['Smoke everyday'] = new_york['Smoke everyday'].str.replace('%', '')
new_york['Smoke everyday'] = pd.to_numeric(new_york['Smoke everyday'])

new_york['Smoke some days'] = new_york['Smoke some days'].str.replace('%', '')
new_york['Smoke some days'] = pd.to_numeric(new_york['Smoke some days'])

new_york['Former smoker'] = new_york['Former smoker'].str.replace('%', '')
new_york['Former smoker'] = pd.to_numeric(new_york['Former smoker'])

new_york['Never smoked'] = new_york['Never smoked'].str.replace('%', '')
new_york['Never smoked'] = pd.to_numeric(new_york['Never smoked'])
```

```python
#as I was looking at this data, I had a lot of questions that I want to find out. This will take looking at the dataset as a whole.
#mainly I'm concerned if smoker's are just moving elsewhere, that's my first concern(or if smoking popularity is nomadic)
df['Smoke everyday'] = df['Smoke everyday'].str.replace('%', '')
df['Smoke everyday'] = pd.to_numeric(df['Smoke everyday'])

df['Smoke some days'] = df['Smoke some days'].str.replace('%', '')
df['Smoke some days'] = pd.to_numeric(df['Smoke some days'])

df['Former smoker'] = df['Former smoker'].str.replace('%', '')
df['Former smoker'] = pd.to_numeric(df['Former smoker'])

df['Never smoked'] = df['Never smoked'].str.replace('%', '')
df['Never smoked'] = pd.to_numeric(df['Never smoked'])

df.head()
```

```python
plt.figure(figsize=(15,30))
sns.lineplot(data = df, x ='Year', y = 'Never smoked',  hue = 'State')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Never Smoked')
plt.show()
```

```python
#I want to try something, get rid of all of the values between 1995 and 2005
df_1996 = df.loc[df['Year'] == 1996]
df_2005 = df.loc[df['Year'] == 2005]
```

```python
df2 = pd.concat([df_1996, df_2005])
```

```python
plt.figure(figsize=(15,20))
sns.lineplot(data = df2, x ='Year', y = 'Never smoked',  hue = 'State')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Never Smoked')
plt.show()
```

```python
plt.figure(figsize=(15,20))
sns.lineplot(data = df2, x ='Year', y = 'Former smoker',  hue = 'State')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Former Smoker')
plt.show()
```

```python
plt.figure(figsize=(15,20))
sns.lineplot(data = df2, x ='Year', y = 'Smoke everyday',  hue = 'State')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Smoke everyday')
plt.show()
```

```python
former_1996 = df.loc[df['Year'] == 1996]['Former smoker'].mean()
former_2005 = df.loc[df['Year'] == 2005]['Former smoker'].mean()
print('Former 1996: {}, Former 2005: {}'.format(former_1996, former_2005))
never_1996 = df.loc[df['Year'] == 1996]['Never smoked'].mean()
never_2005 = df.loc[df['Year'] == 2005]['Never smoked'].mean()
print('Never 1996: {}, Never 2005: {}'.format(never_1996, never_2005))
smokes_1996 = df.loc[df['Year'] == 1996]['Smoke everyday'].mean()
smokes_2005 = df.loc[df['Year'] == 2005]['Smoke everyday'].mean()
print('Smokes 1996: {}, Smokes: 2005: {}'.format(smokes_1996, smokes_2005))
```

```python
#so are we really lowering smoking? has anything changed? I've grabbed some U.S populations from https://www.multpl.com/united-states-population/table/by-year
pop_1996 = 269390000
pop_2005 = 295520000
num_1996_former = round((former_1996 * pop_1996)/100, 2)
num_2005_former = round((former_2005 * pop_2005)/100, 2)
num_1996_never = round((never_1996 * pop_1996)/100, 2)
num_2005_never = round((never_2005 * pop_2005)/100, 2)
num_1996_smokes = round((smokes_1996 * pop_1996)/100, 2)
num_2005_smokes = round((smokes_2005 * pop_2005)/100,2)
print('# of formers in 1996: {}, # of formers in 2005: {}, # of nevers in 1996: {}, # of nevers in 2005: {}'.format(num_1996_former, 
                                                                                                                   num_2005_former,
                                                                                                                   num_1996_never,
                                                                                                                   num_2005_never))
print('# of smokers in 1996: {}, # of smokers in 2005: {}'.format(num_1996_smokes, num_2005_smokes))
```

Nice! It looks like people are indeed releasing smoking according to publically surveyed data in general. We can see from the 1995, 2005 Never smoked graph that many of the U.S states and territories have a positive slope from 1995 - 2005. While we also have many areas with steep declines, I am believing that these are misleading. Looking at the data from the 'Smoke everyday' graph we see that there is convincing visual proof that the population is either no longer smoking or not recording it.
After checking out the populations from the 1996 and 2005 years and checking the percentages, we are seeing that the numbers are promising even though the population is rising. We have a theoretical 7 million less people smoking in 2005 then in 1996 even though the population has risen 26 million. 

I want to point out that this is recorded smoking. Anecdotally I had worked in benefits adminstration and part of the gig is knowing how employers pay for insurance for their employees. I have seen multiple employers be blatant in explaining that they will pay less for premiums for smoker's because longterm these employees are more expensive and also perhaps for incentive to get employees to stop smoking. These incident has two implications: these types of programs may be contributing to the decrease in smoking and these programs may be incentivizing people to deny smoking even though that they do, for fear of having their premiums increase.

I wanted to bring this point up because the bare number of people is large if the percentages are correct.
